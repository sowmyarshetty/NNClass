```python
# Import the SentenceTransformer class and utility function class from the sentence_transformers module
from sentence_transformers import SentenceTransformer, util
# Use the `all-MiniLM-L6-v2` model.
model = SentenceTransformer('all-MiniLM-L6-v2')
# Import pandas
import pandas as pd
pd.set_option('max_colwidth', 200)
```

```python
# Create a DataFrame for the "news_headlines.csv"
df_news_headlines = pd.read_csv("news_headlines.csv")
df_news_headlines.head()
```

```python
# Convert the "headline" column to a list
headline_list = df_news_headlines["headline"].tolist()
headline_list[:5]
```

```python
# Get the vector embeddings for the headlines.
headline_embeddings = model.encode(headline_list)
headline_embeddings[:5]
```

```python
# Get the vector embeddings from the following news headline.
new_headline = "Top 10 Hacks for Traveling Like a Pro."
new_headline_embedding = model.encode([new_headline])
new_headline_embedding
```

```python


# List to store tuples of (news headline, similarity score)
similarities = []

# Loop through the headline embeddings and corresponding headlines
for headline, embedding in zip(headline_list, headline_embeddings):

    # Calculate cosine similarity score
    similarity_score = util.cos_sim(new_headline_embedding, embedding).item()  # Convert tensor to scalar

    # Append (headline, similarity_score) to list
    similarities.append((headline, similarity_score))

# Sort by similarity score in descending order
similarities.sort(key=lambda x: x[1], reverse=True)

# Print the top 5 most similar headlines
for headline, score in similarities[:5]:
    print(f"{headline} - Similarity: {score:.4f}")
```
